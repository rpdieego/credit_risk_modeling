### Data Dictionary

*   **addr_state** - The state provided by the borrower in the loan application;
*   **annual_inc** - The self-reported annual income provided by the borrower during registration;
*   **annual_inc_joint** - The combined self-reported annual income provided by the co-borrowers during registration;

*   **collection_recovery_fee** - post charge off collection fee;
*   **collections_12_mths_ex_med** - Number of collections in 12 months excluding medical collections;
*   **delinq_2yrs** - The number of 30+ days past-due incidences of delinquency in the borrower's credit file for the past 2 years;
*   **desc** - Loan description provided by the borrower;
*   **dti** - A ratio calculated using the borrower’s total monthly debt payments on the total debt obligations, excluding mortgage and the requested LC loan, divided by the borrower’s self-reported monthly income;
*   **dti_joint** - A ratio calculated using the co-borrowers' total monthly payments on the total debt obligations, excluding mortgages and the requested LC loan, divided by the co-borrowers' combined self-reported monthly income;

*   **fico_range_low** - The lower boundary range the borrower’s FICO at loan origination belongs to;

*   **funded_amnt_inv** - The total amount committed by investors for that loan at that point in time;

### Calculating Weight of Evidence for 'Grades'

```
# Get a dataframe Grade | good_bad

df_grades = pd.concat([df_inputs_prepr['grade'], df_targets_prepr], axis = 1)
df_grades.head()
```

```
# Count entries per grade
df_grades.groupby(df_grades.columns.values[0], as_index = False)[df_grades.columns.values[1]].count()
```

```
# Count entries = 1 per grade
df_grades.groupby(df_grades.columns.values[0], as_index = False)[df_grades.columns.values[1]].mean()
```

```
# Concat number of observations and probability of good grade into a dataframe
df_grades_stats = pd.concat([df_grades.groupby(df_grades.columns.values[0], as_index = False)[df_grades.columns.values[1]].count(),
                df_grades.groupby(df_grades.columns.values[0], as_index = False)[df_grades.columns.values[1]].mean()], axis = 1)
# Drop repeated column
df_grades_stats = df_grades_stats.iloc[: , [0,1,3]]
# Add some labels
df_grades_stats.columns = [df_grades.columns.values[0],'n_obs','prop_good']
df_grades_stats.head()
```

```
# Proportion of observations
df_grades_stats['prop_n_obs'] = df_grades_stats['n_obs']/df_grades_stats['n_obs'].sum()
df_grades_stats.head()
# Number of good borrowers
df_grades_stats['n_good'] = df_grades_stats['prop_good'] * df_grades_stats['n_obs']
# Number of bad borrowers
df_grades_stats['n_bad'] = (1 - df_grades_stats['prop_good']) * df_grades_stats['n_obs']
# Proportion of good and bad borrowers per grade
df_grades_stats['prop_n_good'] = df_grades_stats['n_good'] / df_grades_stats['n_good'].sum()
df_grades_stats['prop_n_bad'] = df_grades_stats['n_bad'] / df_grades_stats['n_bad'].sum()

# Calculate Weight of Evidence
df_grades_stats['WoE'] = np.log(df_grades_stats['prop_n_good']/df_grades_stats['prop_n_bad'])

# Sorte by WoE
df_grades_stats = df_grades_stats.sort_values(['WoE'])

# Reset index
df_grades_stats = df_grades_stats.reset_index(drop = True)

# Differences between subsequent proportion of good borrowers and subsequent Woe
df_grades_stats['diff_prop_good'] = df_grades_stats['prop_good'].diff().abs()
df_grades_stats['diff_WoE'] = df_grades_stats['WoE'].diff().abs()

# Calculate Information Value
df_grades_stats['IV'] = (df_grades_stats['prop_n_good'] - df_grades_stats['prop_n_bad']) * df_grades_stats['WoE']
df_grades_stats['IV'] = df_grades_stats['IV'].sum()

'''
IV feature is the same for all the grade categories, as it refers to the
feature 'Grade' in general

'''

df_grades_stats
```

```
def woe_discrete(df, discrete_variable_name, good_bad_variable_df):
  '''

  INPUTS: (df) - dataframe
          (discrete_variable_name) - chosen feature to calculate Information Value
          (good_bad_variable_df) -  target variable

  OUTPUT: (df) -  Processed dataframe

  '''

  df = pd.concat([df[discrete_variable_name],good_bad_variable_df], axis = 1)

  df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                  df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
  
  # Drop repeated column
  df =  df.iloc[:, [0,1,3]]
  # Add some labels
  df.columns = [df.columns.values[0], 'n_obs', 'prop_good']

  # Proportion of observations
  df['prop_n_obs'] = df['n_obs']/df['n_obs'].sum()
  df.head()
  # Number of good borrowers
  df['n_good'] = df['prop_good'] * df['n_obs']
  # Number of bad borrowers
  df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
  # Proportion of good and bad borrowers per grade
  df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
  df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()

  # Calculate Weight of Evidence
  df['WoE'] = np.log(df['prop_n_good']/df['prop_n_bad'])

  # Sorte by WoE
  df = df.sort_values(['WoE'])

  # Reset index
  df = df.reset_index(drop = True)

  # Differences between subsequent proportion of good borrowers and subsequent Woe
  df['diff_prop_good'] = df['prop_good'].diff().abs()
  df['diff_WoE'] = df['WoE'].diff().abs()

  # Calculate Information Value
  df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
  df['IV'] = df['IV'].sum()

  return df
  
```

```
def plot_by_woe(df_WoE, rotation_of_x_axis_labels = 0):
  '''

  '''

  x = np.array(df_WoE.iloc[:,0].apply(str))
  y = df_WoE['WoE']

  plt.figure(figsize = (18,6))
  plt.plot(x,y, marker = 'o', linestyle = '--', color = 'k')
  plt.title('Weight of Evidence by ' + df_WoE.columns[0])
  plt.xlabel(df_WoE.columns[0])
  plt.ylabel('Weight of Evidence')
  plt.xticks(rotation = rotation_of_x_axis_labels)
```

### Preprocessing Continuous Variables: Creating Dummy Variables

```
def woe_ordered_continuous(df, discrete_variable_name, good_bad_variable_df):
  '''

  INPUTS: (df) - dataframe
          (discrete_variable_name) - chosen feature to calculate Information Value
          (good_bad_variable_df) -  target variable

  OUTPUT: (df) -  Processed dataframe

  '''

  df = pd.concat([df[discrete_variable_name],good_bad_variable_df], axis = 1)

  df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                  df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
  
  # Drop repeated column
  df =  df.iloc[:, [0,1,3]]
  # Add some labels
  df.columns = [df.columns.values[0], 'n_obs', 'prop_good']

  # Proportion of observations
  df['prop_n_obs'] = df['n_obs']/df['n_obs'].sum()
  df.head()
  # Number of good borrowers
  df['n_good'] = df['prop_good'] * df['n_obs']
  # Number of bad borrowers
  df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
  # Proportion of good and bad borrowers per grade
  df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
  df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()

  # Calculate Weight of Evidence
  df['WoE'] = np.log(df['prop_n_good']/df['prop_n_bad'])

  # Differences between subsequent proportion of good borrowers and subsequent Woe
  df['diff_prop_good'] = df['prop_good'].diff().abs()
  df['diff_WoE'] = df['WoE'].diff().abs()

  # Calculate Information Value
  df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
  df['IV'] = df['IV'].sum()

  return df
```
